```python
def distinct_points(values, tol=1e-3):
    points = []
    for v in values:
        if not any(abs(v - p) < tol for p in points):
            points.append(v)
    return points

def find_n_cycles(r_values, n):
    results = []
    for r in r_values:
        if len(distinct_points(logistic_equation(r, 0.2, n=1000)[900:])) == n:
            results.append(r)
    return results
```

```python
r_values = np.linspace(3, 4, 10000)

r_2 = find_n_cycles(r_values, 2)[0]
r_4 = find_n_cycles(r_values, 4)[0]
r_8 = find_n_cycles(r_values, 8)[0]
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 1, 1000)
r = 4
x0 = 0.1

def f(x):
    return r*x*(1 - x)

ax.plot(x, f(x), c="k")
ax.plot(x, x, c="k")

iterations = 0
curr_x = x0
ax.plot(curr_x, curr_x)

while iterations < 300:
    next_x = f(curr_x)
    ax.plot([curr_x, curr_x], [curr_x, next_x], c="r", linewidth=0.5)
    ax.plot([curr_x, next_x], [next_x, next_x], c="r", linewidth=0.5)
    curr_x = next_x
    iterations += 1

ax.set_ylim(0, 1)
ax.set_xlim(0, 1)
ax.set_aspect(1)
```
