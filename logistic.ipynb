We plot the equation
\begin{equation}
x_n = rx_{n-1}(1 - x_{n-1})
\end{equation}
for different values of $r$.

```python
import matplotlib.pyplot as plt
import numpy as np
from chaos import logistic_equation, logistic_tail, distinct_points, find_n_cycles, liapunov
```

We first show scatter plots for $r \in \{1, 2, 3, 4\}$.

```python
x = range(5000)

y1 = logistic_equation(1, 0.1, n=5000)
y2 = logistic_equation(2, 0.1, n=5000)
y3 = logistic_equation(3, 0.1, n=5000)
y4 = logistic_equation(4, 0.1, n=5000)

fig, ax = plt.subplots(2, 2, sharex=True)

ax[0, 0].scatter(x, y1, s=0.5, c="blue"); ax[0, 0].set_title("r=1")
ax[0, 1].scatter(x, y2, s=0.5, c="blue"); ax[0, 1].set_title("r=2")
ax[1, 0].scatter(x, y3, s=0.5, c="blue"); ax[1, 0].set_title("r=3")
ax[1, 1].scatter(x, y4, s=0.5, c="blue"); ax[1, 1].set_title("r=4")

plt.show()
```

Here we have plotted cobweb diagrams which are another way of demonstrating the behaviour of the logistic map.

```python
fig, ax = plt.subplots()

x = np.linspace(0, 1, 1000)
r = 4
x0 = 0.1

def f(x):
    return r*x*(1 - x)

ax.plot(x, f(x), c="k")
ax.plot(x, x, c="k")

iterations = 0
curr_x = x0
ax.plot(curr_x, curr_x)

while iterations < 300:
    next_x = f(curr_x)
    ax.plot([curr_x, curr_x], [curr_x, next_x], c="r", linewidth=0.5)
    ax.plot([curr_x, next_x], [next_x, next_x], c="r", linewidth=0.5)
    curr_x = next_x
    iterations += 1

ax.set_ylim(0, 1)
ax.set_xlim(0, 1)
ax.set_aspect(1)

ax.set_xlabel("$x_n$")
ax.set_ylabel("$x_{n+1}$")
```

We test the liapunov function, and then use it to determine when chaos begins.

```python
print(liapunov(2.8, 1000))
print(liapunov(3, 1000))
print(liapunov(3.4, 1000))
print(liapunov(3.9, 1000))
```

```python
r = 3.55
while liapunov(r, 1000) < 0:
        r += 0.00001
print(f"Chaos starts at approx {r}")
        
```

We plot a graph to show the Liapunov exponent for different values of $r$.

```python
r = np.linspace(3, 4, 1000)

fig, ax = plt.subplots()

ax.plot(r, liapunov(r, 10000), 'b-', linewidth=0.8)
ax.plot(r, np.zeros(1000), 'k--', label="$\lambda=0$")
ax.set_xlabel("r", fontsize=20)
ax.set_ylabel("$\lambda$", fontsize=20)
ax.tick_params(labelsize=15)

ax.set_ylim(-1, 1)
ax.legend(fontsize=15)
fig.tight_layout()

fig.savefig("liapunov.pdf")
```
